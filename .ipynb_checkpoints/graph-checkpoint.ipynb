### Flow Chart

```python
# Create a new directed graph
dot = Digraph()

# Define nodes
dot.node('0', label='Prepare Peptoids', shape='box')
dot.node('1', label='Force Fields Choice', shape='box')
dot.node('1.1', label='GAFF1', shape='box')

# Use HTML-like labels for nodes with multiple Other Force Fields
dot.node('1.2', label='<<TABLE BORDER="0" CELLBORDER="1" CELLSPACING="0"><TR><TD>Other Force Fields</TD></TR><TR><TD>GAFF2</TD></TR><TR><TD>OpenFF 2.0</TD></TR><TR><TD>MEFTOID</TD></TR><TR><TD>CGenFF-NTOID</TD></TR></TABLE>>', shape='plaintext')

# Use HTML-like labels for nodes with multiple fields
dot.node('2', label='Prepare Productive Run', shape='box')
dot.node('2.1', label='<<TABLE BORDER="0" CELLBORDER="1" CELLSPACING="0"><TR><TD PORT="EM">EM</TD></TR><TR><TD PORT="NVT">NVT</TD></TR><TR><TD PORT="NPT">NPT</TD></TR></TABLE>>', shape='plaintext')

dot.node('3', 'MD Simulation', shape='box')
dot.node('3.1', 'Hamiltonian Replica Exchange Molecular Dynamic', shape='box')

dot.node('4', 'Analysis', shape='box')
dot.node('4.1', label='<<TABLE BORDER="0" CELLBORDER="1" CELLSPACING="0"><TR><TD PORT="DataCleaning">Data Cleaning</TD></TR><TR><TD PORT="Clustering">Clustering</TD></TR><TR><TD PORT="BICePs">BICePs</TD></TR></TABLE>>', shape='plaintext')

# Define the edges using tuples
edges = [
    ('0', '1'),
    ('1', '1.1'),
    ('1', '1.2'),
    ('1.1', '2'),
    ('2', '2.1'),  # Use the port for the HTML-like label
    ('2', '3'),
    ('3', '3.1'),
    ('3', '4'),
    ('4', '4.1'),  # Use the port for the HTML-like label
]

# Add edges to the graph
for tail, head in edges:
    dot.edge(tail, head)

# Create a subgraph to force horizontal layout for node 2 and 2.1 fields
with dot.subgraph() as s:
    s.attr(rank='same')        
    s.node('1')
    s.node('1.2')
    
with dot.subgraph() as s:
    s.attr(rank='same')      
    s.node('2')
    s.node('2.1')

with dot.subgraph() as s:
    s.attr(rank='same')      
    s.node('3')
    s.node('3.1')

with dot.subgraph() as s:
    s.attr(rank='same')      
    s.node('4')
    s.node('4.1')

# Render the graph
dot.render('flowchart', format='png')
Image(filename='flowchart.png')
```

### Energy Diagram

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the torsion potential function
def torsion_potential(omega, k_phi, n=2, omega_s=0):
    return k_phi * (1 + np.cos(np.radians(n * omega) - np.pi))

# Define the dihedral angle values (e.g., from -50 to 290 degrees)
omega = np.linspace(-50, 290, 1000)

# Define the maximum value for k_phi and the number of steps
k_phi_max = 10.46000
num_steps = 6

# Calculate the lambda values for each step
lambda_values = np.linspace(1, 0, num_steps)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))
for i, l in enumerate(lambda_values):
    k_phi = l * k_phi_max
    V = torsion_potential(omega, k_phi, n=2, omega_s=0)
    color = plt.cm.plasma(i / (num_steps - 1))  # Get color from colormap
    ax.plot(omega, V, label=f'k_phi = {k_phi:.2f}, $\\lambda$ = {l:.2f}', color=color)

# Overlay the image
#img = mpimg.imread('flowchart.png')  # Replace with the path to your image file

# Add labels and title
ax.set_xlabel('Dihedral Angle (ω) [degrees]')
ax.set_ylabel('Potential Energy (V)')
ax.set_title('Torsion Potential Function with Varying Force Constants (Minima at 0° and 180°)')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.grid(False)

# Show the plot
plt.show()
```
