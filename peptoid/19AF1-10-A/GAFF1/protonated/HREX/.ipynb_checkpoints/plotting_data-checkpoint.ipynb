```python
import os
import re
import glob
import natsort
import pymbar
import alchemlyb
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from pymbar.timeseries import detect_equilibration, subsample_correlated_data
from alchemlyb.preprocessing import subsampling
from alchemlyb.estimators import MBAR
from alchemlyb.parsing.gmx import extract_u_nk
from alchemlyb.visualisation import plot_mbar_overlap_matrix

# Just some settings for plotting
rc('font', **{
    'family': 'sans-serif',
    'sans-serif': ['DejaVu Sans'],
    'size': 10,
})

# Set the font used for MathJax - more on this print(images)
rc('mathtext', **{'default': 'regular'})
plt.rc('font', family='serif')
```

```python
# number of replicas
N = 6
# simulation temperature
T = 300

dhdl_dir = 'dhdl_files'

folders = [f'{dhdl_dir}/state_{i}' for i in range(N)]
u_nk_data = []
for i in range(N):
    files = glob.glob(os.path.join(folders[i], '*dhdl*xvg*'))
    files = natsort.natsorted(files, reverse=False)
    print(f' Collecting data for the following files: {", ".join(files)}')
    u_nk = alchemlyb.concat([extract_u_nk(xvg, T=T) for xvg in files])
    u_nk_series = subsampling.u_nk2series(u_nk)
    u_nk, u_nk_series = subsampling._prepare_input(u_nk, u_nk_series, drop_duplicates=True, sort=True)
    u_nk = subsampling.slicing(u_nk, step=1)
    u_nk_series = subsampling.slicing(u_nk_series, step=1)

    t, statinef, Neff_max = detect_equilibration(u_nk_series.values)
    u_nk_series_equil, u_nk_equil = u_nk_series[t:], u_nk[t:]
    indices = subsample_correlated_data(u_nk_series_equil, g=statinef)
    preprocessed_u_nk = u_nk_equil.iloc[indices]
    u_nk_data.append(preprocessed_u_nk)

u_nk_data = alchemlyb.concat(u_nk_data)
print(' Fitting MBAR on u_nk ...')
try:
    mbar_stop = False
    mbar = MBAR().fit(u_nk_data)
except pymbar.utils.ParameterError():
    mbar_stop = True
    print("\sum_n W_nk is not equal to 1, probably due to insufficient overlap between states.")
    print("Stop using MBAR ...")

if mbar_stop is False:
    print(f" MBAR: {mbar.delta_f_.iloc[0, -1]:.4f} +/- {mbar.d_delta_f_.iloc[0, -1]:.4f} kT")
```

```python
plot_mbar_overlap_matrix(mbar.overlap_matrix)
```

```python
import re
import matplotlib.pyplot as plt

# Define the patterns to search for
step_pattern = re.compile(r'Step\s+Time')
proper_dih_pattern = re.compile(r'Bond\s+Angle\s+Proper Dih\.\s+Improper Dih\.\s+LJ-14')
steps = []
proper_dihs = []

# Open the file in read mode
with open('state_1/HREMD.part0001.log', 'r') as file:
    step_found = False  # Initialize a flag to indicate when the pattern is found
    dih_found = False
    line_counter = 0  # Counter to track lines after the pattern
    
    # Read the file line by line
    for line in file:
        if step_found:
            line_counter += 1
            if line_counter == 1:
                # Extract and store the step value (first line after the pattern)
                step = line.strip().split()[0]
                steps.append(step)
                
                step_found = False

        elif step_pattern.search(line):
            # Set the flag and counter when the step pattern is found
            step_found = True
            line_counter = 0

        if dih_found:
            line_counter += 1
            if line_counter == 1:
                # Extract and store the proper dihedral value (first line after the pattern)
                proper_dih = line.strip().split()[2]
                proper_dihs.append(proper_dih)
                
                dih_found = False
            
        elif proper_dih_pattern.search(line):
            # Set the flag and counter when the proper dihedral pattern is found
            dih_found = True
            line_counter = 0

# Print the lengths of the lists
print(f"Number of steps: {len(steps)}")
print(f"Number of proper dihedrals: {len(proper_dihs)}")

# If the lengths are mismatched, handle the discrepancy
if len(steps) != len(proper_dihs):
    print("Warning: Mismatch in the number of steps and proper dihedrals!")
    min_length = min(len(steps), len(proper_dihs))
    steps = steps[:min_length]
    proper_dihs = proper_dihs[:min_length]

# Convert the extracted values to floats for plotting
steps = [float(step) for step in steps]
proper_dihs = [float(proper_dih) for proper_dih in proper_dihs]

# Plotting the values
plt.figure(figsize=(10, 5))
plt.plot(steps, proper_dihs, marker='o', linestyle='-', color='b')
plt.xlabel('Steps')
plt.ylabel('Proper Dihedral Angles')
plt.title('Steps vs Proper Dihedral Angles')
plt.grid(True)
plt.show()
```
